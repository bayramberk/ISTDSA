```python
def dayName(day_of_week):
    if day_of_week==0:
        return 'Monday'
    elif day_of_week==1:
        return 'Tuesday'
    elif day_of_week==2:
        return 'Wednesday'
    elif day_of_week==3:
        return 'Thursday'
    elif day_of_week==4:
        return 'Friday'
    elif day_of_week==5:
        return 'Saturday'
    else:
        return 'Sunday'
```

```python
def timeGroups(time):
    if time >= pd.to_datetime('07:00:00').time() and time <pd.to_datetime('09:00:00').time():
        return '07_09'
    elif time >= pd.to_datetime('09:00:00').time() and time <pd.to_datetime('12:00:00').time():
        return '09_12'
    elif time >= pd.to_datetime('12:00:00').time() and time <pd.to_datetime('14:00:00').time():
        return '12_14'
    elif time >= pd.to_datetime('14:00:00').time() and time <pd.to_datetime('17:00:00').time():
        return '14_17'
    elif time >= pd.to_datetime('17:00:00').time() and time <pd.to_datetime('19:00:00').time():
        return '17_19'
    elif time >= pd.to_datetime('19:00:00').time() and time <pd.to_datetime('23:00:00').time():
        return '19_23'
    else:
        return '23_07'
```

```python
df_summer['time_to_from']=df_summer['time'].apply(timeGroups)
df_summer
```

```python
df_summer.info()
```

```python
df_first_station = df_summer[df_summer['station_complex']=='Jefferson St (L)']
df_first_station.head()
```

```python
df_first_station_grouped_days = df_first_station.groupby(['name_of_day'])[['ridership']].sum().sort_values('ridership', ascending=False)
df_first_station_grouped_days
```

```python
plt.figure(figsize = [15,4])
plt.title("Jefferson St (L)", weight='bold',color="red")
plt.pie(df_first_station_grouped_days['ridership'],labels=df_first_station_grouped_days.index,autopct='%1.1f%%', startangle=95);
```

```python
df_first_station_busiest_day_grouped = df_first_station.groupby(['name_of_day','time_to_from'])[['ridership']].sum().reset_index()
df_first_station_busiest_day_grouped = df_first_station_busiest_day_grouped.pivot_table(index='time_to_from', columns='name_of_day', values='ridership')
df_first_station_busiest_day_grouped
```

```python
plt.figure(figsize = [15,10])
plt.title('Jefferson St (L)',weight = 'bold',color='red',fontsize=15)
plt.xlabel('Days', fontsize=10, weight = 'bold', color='black')
plt.ylabel('To From',fontsize = 10, weight = 'bold', color='black')
plt.xticks(fontsize = 8, weight = 'bold', color='black')
plt.yticks(fontsize = 8, weight = 'bold', color='black')
plt.text(2,6.5, 'Busiest Day and Hour ', weight = 'bold',fontsize = 8, color = 'white'); # Yaz脹n脹n balama yeri(x, y)
sns.heatmap(df_first_station_busiest_day_grouped, cmap='Oranges');
```
